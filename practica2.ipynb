### 3.1. Coeficientes de Fresnel
Cuando la luz se refleja/refracta en una superfice, el haz incidente, reflejado y refractado forman un único plano con la normal. Llamamos a este plano PLANO DE INCIDENCIA. 

Por último, conviene recordar que mediremos los ángulos de los haces incidentes, reflejado refractado respecto a la normal. Cuando el haz incide de manera frontal a la interfase, $\theta_i=0$.

A la hora de estudiar la reflexión (idem refracción), descomponemos el campo de un haz linealmente polarizado en la componente paralela al plano de incidencia ($E_\parallel$) y la componente perpendicular ($E_\perp$). Los subíndices $i$, $r$ y $t$ se refieren al haz incidente, reflejado y refractado (transmitido), respectivamente.

<figure style="text-align: center;">
    <img src="hecht-componentes.png" alt="Componentes paralelas y perpendiculares" style="max-width:50%; height:auto;"/>
    <figcaption> Reflexión y refracción. Imagen tomada de Hecht, Pearson 2017.
</figure>

Tras la reflexión, la amplitud del campo en cada componente sufre un cambio que depende del ángulo de incidencia ($\theta_i$) y los índices de refracción de cada medio ($n_i$, $n_t$). Se puede demostrar (ver apuntes de clase) que estos cambios pueden ser calculados a partir de los llamados *coeficientes de reflexión en amplitudes* de Fresnel:

$$r_\perp \equiv \frac{E_{0r}^\perp}{E_{0i}^\perp} = ... = \frac{n_i \cos\theta_i - n_t \cos\theta_t}{n_i \cos\theta_i + n_t \cos\theta_t}$$

$$r_\parallel \equiv \frac{E_{0r}^\parallel}{E_{0i}^\parallel} = ... = \frac{n_t \cos\theta_i - n_i \cos\theta_t}{n_t \cos\theta_i + n_i \cos\theta_t},$$

donde a su vez el ángulo refractado $\theta_t$ viene determinado a través de la ley de Snell, $n_i \sin\theta_i = n_t \sin\theta_t$.

Generalmente no medimos amplitudes, sino energías o irradiancias. En este caso, se puede demostrar que la fracción de potencia reflejada viene dada por el coeficiente de reflexión en energías,
$$R \equiv \frac{\mathrm{potencia\;reflejada}}{\mathrm{potencia\;incidente}} = \frac{I_r}{I_i} = r^2,$$
la cual vale para un haz linealmente polarizado en el plano de incidencia ($R_\parallel = r_\parallel^2$) o perpendicular a este ($R_\perp = r_\perp^2$).

* **NOTA**: para el caso de la reflexión, es equivalente trabajar con potencias o irradiancias (estas últimas son las que medimos con el polímetro). Si quisiéramos estudiar el coeficiente de transmisión para el haz refractado, habría que hacerlo con más cuidado (ver apuntes de clase).

### 3.2. Ángulo de Brewster
A la hora de estudiar los coeficientes, se puede ver que existe un ángulo de incidencia no trivial, $\theta_i = \theta_B$, para el cual $r_\parallel=0$, $r_\perp\neq0$ (y no al revés). Esto tiene varias implicaciones:
* Si hacemos incidir luz linealmente polarizada en el plano de incidencia, no existe haz reflejado, y toda la luz se transmite.

* Si hacemos incidir luz natural (que podemos siempre descomponer en componentes $\parallel$ y $\perp$) con ese ángulo, la luz reflejada solo tiene componente perpendicular, es decir, se refleja polarizada.

<figure style="text-align: center;">
    <img src="brewsters-angle.svg" alt="Ángulo de Brewster" style="max-width:50%; height:auto;"/>
    <figcaption> Polarización por reflexión en el ángulo de Brewster. Imagen tomada de Wikipedia.
</figure>

Se puede demostrar (ver apuntes de clase) que este ángulo es:
$$\theta_B = \tan^{-1}(n_t/n_i).$$

### 3.3. Ángulo azimutal
Los coeficientes de reflexión en amplitudes de Fresnel nos dicen que las componentes paralelas y perpendiculares de un haz no necesariamente se reflejan en la misma proporción. Esto tiene un efecto directo: *el plano de oscilación puede cambiar tras una reflexión*.

El **ángulo azimutal** es el ángulo que forma el plano de oscilación con el plano de incidencia. Llamremos $\alpha$ al ángulo azimutal del haz incidente, y $\rho$ al ángulo azimutal del haz reflejado. Se puede demostrar que:

$$\tan\rho = \frac{r_\perp}{r_\parallel} \tan\alpha = ... = -\frac{\cos(\theta_i-\theta_t)}{\cos(\theta_i+\theta_t)}\tan\alpha.$$

* **NOTA:** Esta fórmula predice que, para incidencia normal ($\theta_i = \theta_t=0$), hay un cambio de signo en el ángulo azimutal, $\rho=-\alpha$.
