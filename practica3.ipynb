## 3. Fundamento terico

La figura representa un haz de luz monocromtico (longitud de onda $\lambda$) que se refracta dos veces en las caras de un prisma (sumergido en aire). Llamamos:
* $A$: *ngulo de apertura* del prisma (60 para un prisma equiltero).
* $\theta_1$ y $\theta_2$: ngulos externos respecto a las normales (incidencia y refractado).
* $\theta_1'$ y $\theta_2'$: ngulos internos respecto a las normales (refractado e incidencia).
* $D$: ngulo de desviacin entre la direccin del haz de salida y el incidente.



Usando geometra bsica, se puede ver que estos ngulos cumplen algunas relaciones sencillas (ver apuntes de clase):
$$ A = \theta_1'+\theta_2'\\
D = \theta_1 + \theta_2-A.$$

Adems, en cada refraccin se cumple la ley de Snell:
$$ 1\cdot \sin\theta_1 = n \sin\theta_1'\\
 1\cdot \sin\theta_2 = n \sin\theta_2'$$

Por ltimo, en medios dispersivos (como el vidrio), el ndice de refraccin vara ligeramente con cada longitud de onda. En primera aproximacin, esta dependencia viene dada por la **ley de Cauchy**:
$$n(\lambda) = B+\frac{C}{\lambda^2},$$
donde $B$ y $C$ son constantes caractersticas del material. 

Para caracterizar el grado de dispersin de un material, utilizamos el llamado **nmero de Abbe**:
$$V \equiv \frac{n_D-1}{n_F-n_C},$$
donde $n_X$ representa el ndice de refraccin del material para tres longitudes de onda caractersticas del espectro visible, que por convenio son:
* C: lnea espectral **roja** del espectro de emisin del hidrgeno ($\lambda_C = 656.3$ nm).
* D: lnea espectral **amarilla** del espectro de emisin del sodio ($\lambda_D = 589.3$ nm). Alternativamente, a veces se usa la lnea espectral amarilla del helio ($\lambda_d = 587.6$ nm).
* F: lnea espectral **azul** del espectro de emisin del hidrgeno ($\lambda_F = 486.1$ nm).
