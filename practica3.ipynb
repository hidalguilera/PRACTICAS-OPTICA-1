## 3. Fundamento teórico

La figura representa un haz de luz monocromático (longitud de onda $\lambda$) que se refracta dos veces en las caras de un prisma (sumergido en aire). Llamamos:
* $A$: *ángulo de apertura* del prisma (60º para un prisma equilátero).
* $\theta_1$ y $\theta_2$: ángulos externos respecto a las normales (incidencia y refractado).
* $\theta_1'$ y $\theta_2'$: ángulos internos respecto a las normales (refractado e incidencia).
* $D$: ángulo de desviación entre la dirección del haz de salida y el incidente.



Usando geometría básica, se puede ver que estos ángulos cumplen algunas relaciones sencillas (ver apuntes de clase):
$$ A = \theta_1'+\theta_2'\\
D = \theta_1 + \theta_2-A.$$

Además, en cada refracción se cumple la ley de Snell:
$$ 1\cdot \sin\theta_1 = n \sin\theta_1'\\
 1\cdot \sin\theta_2 = n \sin\theta_2'$$

Por último, en medios dispersivos (como el vidrio), el índice de refracción varía ligeramente con cada longitud de onda. En primera aproximación, esta dependencia viene dada por la **ley de Cauchy**:
$$n(\lambda) = B+\frac{C}{\lambda^2},$$
donde $B$ y $C$ son constantes características del material. 

Para caracterizar el grado de dispersión de un material, utilizamos el llamado **número de Abbe**:
$$V \equiv \frac{n_D-1}{n_F-n_C},$$
donde $n_X$ representa el índice de refracción del material para tres longitudes de onda características del espectro visible, que por convenio son:
* C: línea espectral **roja** del espectro de emisión del hidrógeno ($\lambda_C = 656.3$ nm).
* D: línea espectral **amarilla** del espectro de emisión del sodio ($\lambda_D = 589.3$ nm). Alternativamente, a veces se usa la línea espectral amarilla del helio ($\lambda_d = 587.6$ nm).
* F: línea espectral **azul** del espectro de emisión del hidrógeno ($\lambda_F = 486.1$ nm).

### 3.1. Método del ángulo de desviación mínima

* Ve a https://phet.colorado.edu/sims/html/bending-light/latest/bending-light_all.html?locale=es (Prismas) y establece un montaje similar al de la figura, en el que el ángulo de incidencia sobre la primera cara del prisma es prácticamente 90º (rasante):



* No olvides seleccionar la caja de "Normal" (abajo a la derecha) para una mejor visualización de los ángulos.

* Gira LENTAMENTE el prisma de forma que el ángulo de incidencia $\theta_1$ sea cada vez más pequeño (gira en sentido antihorario) y hasta que el haz saliente salga prácticamente rasante ($\theta_2\rightarrow 90º$). ¿Qué ha ocurrido con el ángulo de desviación $D$? ¿Aumenta o disminuye?

* Si has observado bien, el ángulo de desviación tiene un MÍNIMO para una determinada orientación. Justo en esa posición, se cumple además que los ángulos externos e internos son iguales entre sí (ver apuntes de clase):

$$\theta_1=\theta_2,\qquad \theta_1'=\theta_2'.$$ 

* Usando estas identidades en las ecuaciones de la sección anterior, se demuestra fácilmente que (puedes probar a hacerlo):

$$\boxed{n = \frac{\sin\left(\frac{D_{\min}+A}{2}\right)}{\sin\left(\frac{A}{2}\right)}.}$$

**NOTA: esta ecuación solo es aplicable si colocamos el prisma en la posición de desviacion mínima**.

La expresión anterior es muy útil, porque permite determinar el índice de refracción del prisma sin tener que medir ángulos de indicencia (lo cual implicaría dibujar normales, etc.), y solo necesita el ángulo de desviación *mínimo* y el ángulo de apertura del prisma.

## 5. Mediciones
### 5.1. Calibración y método de desviación mínima
1. Coloca la lámpara de He deleante del columador, cerca de la apertura.
2. Situa el anteojo **en línea** con el dispositivo colimador hasta observar el haz de luz. Ajusta la apertura y el enfoque para obtener una línea delgada y nítida.
3. Centra la línea en el eje verticual del retículo.
4. Gira la plataforma con el prisma hasta conseguir un ángulo de incidencia sobre la primera cara de casi 90º (rasante). 
5. Mueve el anteojo hasta localizar las líneas del espectro en el haz que emerge del prisma. Vuelve a enfocar si no se veían con suficiente nitidez.
6. **A partir de aquí, deberás girar lentamente el prisma (mientras observas el anteojo) hasta minimizar la desviación.** Una vez alcances esta situación, no muevas más el primsa.

* **NOTAS:**
    * La plataforma permite anotar la posición angular de los haces emergentes (relativa a un cero colocado de forma arbitraria). Al restar la posición del haz sin desviar, la cual es siempre la misma, puedes determinar el ángulo de desviación.

    * Si estás trabajando en la posición de desviación mínima, la ecuación que vimos en la sección anterior te permite determinar $n(D_{\min})$ para cada línea espectral.
