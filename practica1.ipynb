## 3. Fundamento teórico
### 3.1. Ley de Malus
Cuando luz no polarizada (natural, parcialmente polarizada) atraviesa un filtro polarizador, algunas componentes se absorben y la luz que emerge está linealmente polarizada. Si ahora este haz linealmente polarizado, de intensidad $I_0$, se hace pasar por un segundo filtro polarizador (el "analizador") cuyo eje de trasmisión está a un ángulo $\phi$ respecto al primero, la intensidad emergente $I_e$ será:
$$ I_e = I_0 \cos\phi^2.$$

### 3.2. Actividad óptica
Algunas disoluciones tienen la capacidad de rotar el plano de polarización de un haz linealmente polarizado. Este fenómeno se conoce como actividad óptica, y se rige por la ley de Biot:
$$ \alpha = \Gamma l c,$$
siendo $\alpha$ el ángulo rotado, $l$ la distancia que recorre el haz en la disolución y $c$ la concentración de esta. $\Gamma$ es la *constante de rotación específica*, y depende de la disolución y de la propia longitud de onda, de forma que $\Gamma \lambda^2 = \mathrm{cte.}$ Para la longitud de onda del sodio ($\lambda_D = 589.3$ nm,  $\Gamma_D = +66.5^\circ\cdot\mathrm{L/m\cdot g}$ ).

### 5.3. Sacarimetría.

1. Prepare una disolución de azúcar en agua de unos 50-70 g/L. Anote la concentración usada. 

2. Coloque un polarizador delante del láser para asegurar que esté linealmente polarizado en una dirección determinada. 

3. Utilizará el segundo polarizador y el fotómetro como analizador del haz tras pasar por la disolución. 

4. Utilice 3 cubetas de diferente longitud (ancho y largo) y diseñe un experimento para comprobar la ley de Biot ($\alpha\propto l$). 

5. A partir del ajuste lineal, determine la concentración, y compare con el valor que tenía anotado.

* **NOTAS:**
    * Asegúrese de no saturar la disolución, y que toda el azúcar está disuelta. Verterá esta disolución de una cubeta a otra durante el experimento. 

    * A la hora de usar el segundo polarizador (analizador), puede trabajar en intensidad máxima, o bien en intensidad mínima (y luego rotar $90^\circ$ para determinar el estado de polarización). Este último método, a pesar de involucrar un paso más, tiene una ventaja. Discuta cuál.

    * Para deteminar la concentración, necesitará conocer la longitud de onda del láser de He-Ne. Recuerde que para la longitud de onda del sodio, $\lambda_D = 589.3$ nm,  $\Gamma_D = +66.5^\circ\cdot\mathrm{L/m\cdot g}$.
