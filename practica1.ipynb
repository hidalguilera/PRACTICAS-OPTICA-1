## 3. Fundamento teórico
### 3.1. Ley de Malus
Cuando luz no polarizada (natural o parcialmente polarizada) atraviesa un filtro polarizador, algunas componentes se absorben y la luz que emerge está linealmente polarizada. Si ahora este haz linealmente polarizado, de intensidad $I_0$, pasa por un segundo filtro polarizador (el "analizador") cuyo eje de trasmisión está a un ángulo $\phi$ respecto al primero, la intensidad emergente, $I_e$, será:
$$ I_e = I_0 \cos\phi^2.$$

### 3.2. Actividad óptica
Algunas disoluciones tienen la capacidad de rotar el plano de polarización de un haz linealmente polarizado. Este fenómeno se conoce como actividad óptica, y se rige por la ley de Biot:
$$ \alpha = \Gamma l c,$$
donde $\alpha$ es el ángulo rotado, $l$ la distancia que recorre el haz en la disolución y $c$ la concentración de esta. $\Gamma$ es la *constante de rotación específica*, y depende de la disolución y la temperatura. También varía con la propia longitud de onda del haz, de forma que $\Gamma \lambda^2 = \mathrm{cte.}$ Para la longitud de onda del sodio ($\lambda_D = 589.3$ nm) y a una temperatura $20^\circ$,  $\Gamma_D = +66.5^\circ\cdot\mathrm{L/m\cdot g}$.

4. Repite la observación con una regla de plástico o un transportador de ángulos (no es necesario aplicar tensión aquí).
