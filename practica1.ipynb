## 3. Fundamento terico
### 3.1. Ley de Malus
Cuando luz no polarizada (natural o parcialmente polarizada) atraviesa un filtro polarizador, algunas componentes se absorben y la luz que emerge est linealmente polarizada. Si ahora este haz linealmente polarizado, de intensidad $I_0$, pasa por un segundo filtro polarizador (el "analizador") cuyo eje de trasmisin est a un ngulo $\phi$ respecto al primero, la intensidad emergente, $I_e$, ser:
$$ I_e = I_0 \cos\phi^2.$$

### 3.2. Actividad ptica
Algunas disoluciones tienen la capacidad de rotar el plano de polarizacin de un haz linealmente polarizado. Este fenmeno se conoce como actividad ptica, y se rige por la ley de Biot:
$$ \alpha = \Gamma l c,$$
donde $\alpha$ es el ngulo rotado, $l$ la distancia que recorre el haz en la disolucin y $c$ la concentracin de esta. $\Gamma$ es la *constante de rotacin especfica*, y depende de la disolucin y la temperatura. Tambin vara con la propia longitud de onda del haz, de forma que $\Gamma \lambda^2 = \mathrm{cte.}$ Para la longitud de onda del sodio ($\lambda_D = 589.3$ nm) y a una temperatura $20^\circ$,  $\Gamma_D = +66.5^\circ\cdot\mathrm{L/m\cdot g}$.
